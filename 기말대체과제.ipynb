```python
a = [5280000,5501000,5469000,5480000,5533000,5554000]
b = [5280000,5500000,5729000,5968000,6217000, 6476000]
month = ['sep','oct','nov','dec','jan','fed']

df = pd.DataFrame({'총매출':a, '목표매출':b}, index = month)
df
```

```python
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
fig, ax = plt.subplots(figsize=(12,6))
bar_width = 0.25
index = np.arange(6)
b1 = plt.bar(index, df['총매출'], bar_width, alpha=0.4, color='red', label='총매출')
b2 = plt.bar(index + bar_width, df['목표매출'], bar_width, alpha=0.4, color='blue', label='목표매출')
plt.xticks(np.arange(bar_width, 6 + bar_width, 1), month)
plt.xlabel('달', size = 13)
plt.ylabel('단위 : $', size = 13)
plt.legend()
plt.show()

```
